**Matplotlib Plotting Data**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoints1 = np.array([0,2,4,6,8])
ypoints1 = np.array([0,2,9,3,5])

xpoints2 = np.array([0, 2, 4, 6, 8])
ypoints2 = np.array([0, 4, 2, 3, 9])

ypoints3 = np.array([4,6,3,4,2]).cumsum()

ypoints1 = ypoints1.cumsum()
ypoints2 = ypoints2.cumsum()

plt.plot(xpoints1, ypoints1)
plt.plot(xpoints2, ypoints2, 'o--r') # marker | line | color
plt.plot(ypoints3, marker = '*', ms = 20, mec = 'g', mfc = 'b') # accepts hexa
plt.show()
```

```python
plt.plot(xpoints1, ypoints1, xpoints2, ypoints2, linestyle = 'dotted', linewidth='10')

font1 = {'family':'serif', 'color':'blue', 'size': 20}

plt.title("Data Plot Title", fontdict = font1, loc = 'right')
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")

plt.grid(axis = 'y', color = 'green', linestyle= '--', linewidth='0.5')

plt.show()
```

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 9, 2, 6])

plt.subplot(1,2,1) # number of rows and columns
plt.title("graph 1")
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([7, 2, 1, 4])

plt.subplot(1,2,2)
plt.title("graph 2")
plt.plot(x,y)

plt.suptitle("All the Graphs")


plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x,y, color = 'red')

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 65, 70, 75, 80, 90, 100])
sizes = np.array([20,50,100,200,500, 130, 45, 1000,60,90,10,300,600,800,75])

plt.scatter(x, y, c = colors, cmap = 'viridis', s = sizes, alpha = 0.5)

plt.colorbar()
plt.show()
```

```python
x = np.array(["A", "B", "C"])
y = np.array([3, 8, 5])

plt.subplot(1,2,1)
plt.bar(x,y, width = 0.5)

plt.subplot(1,2,2)
plt.barh(x,y, color = 'red', height = 0.1)
plt.show()
```

```python
x = np.random.normal(50, 10, 100)

plt.hist(x, bins = 20, edgecolor='red')
plt.show()
```

```python
y = np.array([36, 25, 25, 14])

labels = ["Apples", "Bread", "Pizza", "Rizz"]
explode = [0.3, 0, 0 ,0]
colors = ["blue", "red", "green", "orange"]

plt.pie(y, colors=colors, labels=labels, startangle = 90, explode=explode, shadow = True)

plt.legend(title = "Foods", loc = 'lower right')
plt.show()
```
